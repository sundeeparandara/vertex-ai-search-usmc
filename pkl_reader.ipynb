```python
import pickle
```

```python
PKL_PATH = "./partitioned_output.pkl"  # Where we save the output
```

```python
def load_elements_pickle(filename):
    with open(filename, "rb") as f:
        elements = pickle.load(f)
    print(f"📂 Loaded cached elements from: {filename}")
    return elements
```

```python
elements = load_elements_pickle(PKL_PATH)
```

```python
for i, el in enumerate(elements[:300]):
    print(f"\n--- Element {i+1} ---")
    # print(f"Type: {type(el).__name__}")
    print(f"Text: {getattr(el, 'text', '')[:300]}")  # Preview first 300 chars
    # print(f"Metadata: {el.metadata.to_dict()}")
```
